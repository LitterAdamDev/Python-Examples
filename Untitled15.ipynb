```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
# Generating a dummy dataset
np.random.seed(0)
latitudes = np.random.uniform(low=-90.0, high=90.0, size=100)
longitudes = np.random.uniform(low=-180.0, high=180.0, size=100)
data = pd.DataFrame({'Latitude': latitudes, 'Longitude': longitudes})
data[:10]
```

```python
# Using DBSCAN for clustering
# The eps and min_samples parameters may need to be adjusted based on the specific dataset
km_per_radian = 6371.0088
epsilon = 1.5 / km_per_radian  # 1.5 km in radians
db = DBSCAN(eps=epsilon, min_samples=5, algorithm='ball_tree', metric='haversine')
data_rad = np.radians(data[['Latitude', 'Longitude']])
db.fit(data_rad)
data['Cluster'] = db.labels_

# Converting to GeoDataFrame
gdf = gpd.GeoDataFrame(data, geometry=[Point(xy) for xy in zip(data.Longitude, data.Latitude)])
gdf[:5]
```

```python
# Calculating midpoint of each cluster
def calculate_midpoint(points):
    x = y = z = 0
    for point in points:
        latitude, longitude = map(np.radians, [point.y, point.x])
        x += np.cos(latitude) * np.cos(longitude)
        y += np.cos(latitude) * np.sin(longitude)
        z += np.sin(latitude)

    x /= len(points)
    y /= len(points)
    z /= len(points)

    longitude = np.arctan2(y, x)
    hypotenuse = np.sqrt(x * x + y * y)
    latitude = np.arctan2(z, hypotenuse)
    return Point(np.degrees(longitude), np.degrees(latitude))

midpoints = gdf.groupby('Cluster')['geometry'].apply(calculate_midpoint).reset_index()
midpoints = gpd.GeoDataFrame(midpoints, geometry='geometry')
midpoints[:5]
```

```python
# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='white', edgecolor='black')
gdf.plot(ax=ax, marker='o', color='blue', markersize=5)
midpoints.plot(ax=ax, marker='x', color='red', markersize=100)
plt.show()
```
